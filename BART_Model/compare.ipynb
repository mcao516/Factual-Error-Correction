```python
preds_path = '../datasets/FactCC/val/val.source'
preds_corrected_path = 'preds/factcc_val_preds_cpbest.hypo'
```

```python
preds = []
with open(preds_path, 'r', encoding='utf-8') as f:
    for line in f:
        line = line.strip()
        preds.append(line)
```

```python
preds_corrected = []
with open(preds_corrected_path, 'r', encoding='utf-8') as f:
    for line in f:
        line = line.strip()
        preds_corrected.append(line)
```

```python
assert len(preds) == len(preds_corrected)
```

```python
diff_counter = 0

for p, pc in zip(preds, preds_corrected):
    p = p[: p.find('</s>') - 1]
    if p != pc and p != pc[1:] and p.lower() != pc.lower():
        diff_counter += 1
        print('- predictions:')
        print(p)
        print('- corrected predictions:')
        print(pc)
        print()
```
